```python
# Formatting and merging the CSV reclassified output from FAM algorithm with Geo-spatial farm field boundaries (DBF file)
#  for the purpose of creating pivot and summary tables, figures, and maps, etc.
#
# Created by Carolyn Rosevelt, NASA-Ames/CSUMB Cooperative 02/14/18

# Import libraries and read-in FAM output csv file.
import pandas as pd
from pandas import read_csv
import numpy as np

# Read-in csv file.
df = read_csv(r"\...\reclassified_YEAR.csv") 

# Calls and prints (displays) dataframe (rows and columns), the contents of the csv file.
df
```

```python
# Relabels and assigns appropriate column headers to dataframe.
df.columns = ['SIMS_ID', 'F20160720','F20160524','F20160422','F20160320','','','','','','','','','S20160720','S20160524','S20160422','S20160320']
```

```python
df['S20160524'].replace([4],[10],inplace=True) # Replaces values of Emergent code '4' for end of winter with Idle code 10.
df['S20160524'].replace([8],[2],inplace=True) # Replaces values of Grains code '8' for end of winter with Crop code '2'.
df['S20160524'].replace([16],[10],inplace=True) # Replaces values of No Crop Yet code '16' for winter with Idle code '10'.
df['S20160720'].replace([4],[10],inplace=True) # Replaces values of Emergent code '4' for summer with Idle code '10'.
df['S20160720'].replace([15],[10],inplace=True) # Replaces values of Emergent code '4' for summer with Idle code '10'.
df['S20160720'].replace([8],[2],inplace=True) # Replaces values of Emergent code '4' for summer with Idle code '10'.

df['S20160524'].replace([0],['NAN'],inplace=True) # Replaces values of '0' for winter with 'NAN', no data.
df['S20160720'].replace([0],['NAN'],inplace=True) # Replaces values of '0' for summer with 'NAN', no data.

# Double check your work. Print out the changes you just made.
df
```

```python
# Following df.loc[(df[columnheader])] similarly use theory of conditional arguments to populate the'YTDxxx" column. 
# Reclassified code 2 == Cropped and code 10 == Idle.
df['YTD20160720'] = '10'
df
df.loc[(df['S20160720'] > 10) & (df['S20160524'] > 10), 'YTD20160720'] = '10'
df
df.loc[df['S20160720'] < 10,'YTD20160720'] = '2'
df
df.loc[df['S20160524'] < 10,'YTD20160720'] = '2'
df
df.loc[df['S20160720'] == "NAN",'YTD20160720'] = "NAN"
df
```

```python
# Names new file and saves the above edits to a brand new csv file.
df.to_csv(r"\...\reclassified_YEAR_formatted.csv", index = False)
```

```python
# Merge newly updated csv with dbf (csv) of the YEAR basemap
import pandas as pd
from pandas import read_csv
a = pd.read_csv(r"\...\reclassified_YEAR_formatted.csv")
b = pd.read_csv(r"\...\base16_ca_poly_170619csv.csv")

merged = a.merge(b, on='SIMS_ID')
merged.to_csv(r"\...\reclass_w_base16csv_merged.csv", index=False)
```
